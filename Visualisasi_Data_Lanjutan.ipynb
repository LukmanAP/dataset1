```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Mengumpulkan Data bedasarkan Tahun dan Jenis Kelamin, disini kita menggunkan contoh data dari tahun 2020 unutk jenis kelamin L dan P, dan kita coba visualisasikannya menggunakan matplotlib menjadikannya sebagai Bar Chart

```python
penduduk = pd.read_csv("Data Penduduk.csv")
tahun2020 = penduduk.query("tahun == 2020")

#Mengumpulkan data jumlah Pria di tahun 2020
data_pria_2020 = tahun2020.query("kelamin == 'L'")

#Mengumpulkan data jumlah Wanita di tahun 2020
data_wanita_2020 = tahun2020.query("kelamin == 'P'")

```

```python
data_pria_2020.head()
```

```python
data_wanita_2020.head()
```

**Bar** **Chart** dan **pie** **chart** Jumlah Penduduk Pria di tahun 2020

```python
propinsi_L = data_pria_2020.loc[:,'propinsi'] 
jumlah_L = data_pria_2020.loc[:,'jumlah']  #Memfilter data dari Jumlah Pria
```

```python
plt.figure(figsize=(10,10))
plt.barh(propinsi_L, jumlah_L)
plt.show()
```

```python
plt.figure(figsize=(15,15))
ax = plt.axes()
ax.pie(jumlah_L, labels = propinsi_L)
```

**Bar** **Chart** dan **pie** **chart** Jumlah Penduduk Wanita di Tahun 2020

```python
propinsi_P = data_wanita_2020.loc[:,'propinsi']
jumlah_P = data_wanita_2020.loc[:,'jumlah'] #Memfilter data dari Jumlah Wanita
```

```python
plt.figure(figsize=(10,10))
plt.barh(propinsi_P, jumlah_P)
plt.show()
```

```python
plt.figure(figsize=(15,15))
ax = plt.axes()
ax.pie(jumlah_P, labels = propinsi_P)
```
