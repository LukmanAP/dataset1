Import Library yang di gunakan

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

Menyiapkan data dan memanggil dataset

```python
dataset = pd.read_csv('Salary.csv')
dataset.keys()
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

Konversi ke data Array

```python
x = np.asarray(dataset)
print(x)
```

Menampilkan data dalam bentuk scatter plot

```python
plt.scatter(x[:,0], x[:,1], label = 'True Position')
plt.xlabel("years experience")
plt.ylabel("salary")
plt.title("Grafik penyebaran Gaji")
plt.show()
```

Mengaktifkan K-Means denagn jumlah K=2

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
```

Menampilkan Nilai Centroid yang di generate oleh algoritma

```python
print(kmeans.cluster_centers_)
```

Memvisualisasikan Hasil Klasterisasi

```python
plt.scatter(x[:,0], x[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("years experience")
plt.ylabel("Gaji")
plt.title("Grafik Hasil Klaterisasi data gaji")
plt.show()
```

Mmvisualisasikan hasl klaterisasi dengan centroid dari masing-masing klaster

```python
plt.scatter(x[:,0], x[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("years experience")
plt.ylabel("Gaji")
plt.title("Grafik Hasil Klaterisasi data gaji")
plt.show()
```
